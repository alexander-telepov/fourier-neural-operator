```python
import scipy.io
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np
import numpy as np
pred = np.load('prediction_0001.npy')
true = np.load('groundtruth_0001.npy')
```

```python
pred=pred.reshape(1,64,64,40)
true=true.reshape(1,64,64,40)
```

```python
pred.shape,true.shape
```

```python
u_pred = np.concatenate((true,pred), axis=2)
u_pred = np.concatenate((u_pred,abs(true-pred)), axis=2)
```

```python
l=40
```

```python
ims = []
fig = plt.figure(figsize=(18,9))
for i in range(l):
    im = plt.pcolormesh(u_pred[0,:, :, i])
    ims.append([im])
    plt.plot([64, 64], [64, 0], 'k-', linewidth=1)
   
#plt.title('u (left) pred (right)')
plt.axis('off')
ani = animation.ArtistAnimation(fig, ims, interval=50, blit=True,
                                repeat_delay=0)

ani.save('u+pred+diff.gif', dpi=200, writer='imagemagick', fps=1)

plt.show()
```
